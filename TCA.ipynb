```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # pyplot is used in plotting graph
import seaborn as sns

df = pd.read_csv('Customer Churn.csv') # df = dataframe
df.head() # head is used in visualize top 5 rows
```

#### replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
def conv(value):
    if value == 1:
        return "yes"
    else:
        return "no"
        
df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

##### converted 0 and 1 Values of senior citizen to yes/no to make easier to understand

```python
ax = sns.countplot(x ='Churn', data = df, palette = ['Orange','Blue']) # palette is used by the differnt color in different different bar.... color is used by single color in all bar

ax.bar_label(ax.containers[0]) # visualize the value
plt.title("Count of Customer by Churn")
plt.show()
```

```python
plt.figure(figsize = (3,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%",colors = ['Blue','Red'])
plt.title("Percentage of Churned Customers", fontsize = 10,color = "blue")
plt.show()
```

##### from the given pie chart we can conclude that 26.54% of our customers have churned out.  # not let's explore the reason behind it

```python
plt.figure(figsize = (5,4))
sns.countplot(x = "gender",data = df, palette = ['orange','Blue'], hue = "Churn") # x = x-axis = gender and hue is used by multiple categories
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(x = "SeniorCitizen",data = df, color = 'Blue') # x = x-axis = gender 
ax.bar_label (ax.containers[0])
plt.title("Count of Customers by SeniorCitizen")
plt.show()
```

```python
grouped = df.groupby(['SeniorCitizen', 'Churn']).size().unstack()
percentages = grouped.div(grouped.sum(axis=1), axis=0) * 100

# Plot the stacked bar chart
plt.figure(figsize=(4, 4))
bottoms = [0] * len(percentages)

# Plot bars for each category in Churn (Yes/No)
for col, color in zip(percentages.columns, ['orange', 'blue']):
    plt.bar(percentages.index, percentages[col], bottom=bottoms, label=col, color=color)
    bottoms = [bottom + value for bottom, value in zip(bottoms, percentages[col])]

# Add percentage labels to each segment
for idx, senior in enumerate(percentages.index):
    running_bottom = 0  # This tracks the bottom for each stacked segment
    for col, color in zip(percentages.columns, ['orange', 'blue']):
        value = percentages.loc[senior, col]
        if value > 0:  # Only label if the value is non-zero
            plt.text(idx, running_bottom + value / 2, f"{value:.1f}%", ha='center', va='center', color='white', fontsize=8)
        running_bottom += value  # Update the position for the next segment

# Customize the plot
plt.title("Churn by SeniorCitizen (Stacked Bar Chart)", fontsize=10)
plt.xlabel("SeniorCitizen")
plt.ylabel("Percentage")
plt.legend(title="Churn", loc="upper left", bbox_to_anchor=(0.9,0.9)) # loc and bbox_to_anchor is used by the location of legend(no,yes)
plt.show()
```

###### Comparative a Greater Percentage of pepole in senior citizen category have churned

```python
# list of columns
columns = ['PhoneService', 'MultipleLines', 'InternetService', 
           'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
           'TechSupport', 'StreamingTV', 'StreamingMovies']

# Setting up the subplot grid
n_rows, n_cols = 3, 3
fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 12))
axes = axes.flatten() # converts a 2D array of subplot axes into a 1D array

# Creating a countplot for each column
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue = df["Churn"])
    axes[i].set_title(f'count plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

# Removing extra axes if any
for j in range(len(columns), len(axes)):
    fig.delaxes(axes[j]) # Remove unused axes

plt.tight_layout()
plt.show()

```

##### This visualization presents countplots for various telecom service attributes, categorized by customer churn (Yes/No). It highlights trends like high churn among customers with specific InternetService or lacking OnlineSecurity, OnlineBackup, and TechSupport. Patterns vary across services, offering insights into churn factors.
